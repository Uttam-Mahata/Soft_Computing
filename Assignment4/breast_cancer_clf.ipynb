## Data Preprocessing

### **1. Handle Missing Data**
We will use **SimpleImputer** from `sklearn` to impute missing values for categorical variables, such as `node-caps` and `breast-quad`, with the most frequent value.

### **2. Encode Categorical Features**
We will use **One-Hot Encoding** for categorical features such as `age`, `menopause`, `tumor-size`, etc. This will convert categorical values into numerical binary columns.

### **3. Scale Numerical Features**
We'll use **StandardScaler** to normalize the integer values, such as `deg-malig`, to ensure all features are on the same scale.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score

# Fetch dataset
breast_cancer = fetch_ucirepo(id=14)

# Extract features and target
X = breast_cancer.data.features
y = breast_cancer.data.targets

# One-hot encode the target for binary classification
y = np.where(y == 'no-recurrence-events', 0, 1)

# Preprocessing (One-hot encode categorical features, standardize numeric)
categorical_features = ['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'breast', 'breast-quad', 'irradiat']
numeric_features = ['deg-malig']

encoder = OneHotEncoder(handle_unknown='ignore')
X_categorical = encoder.fit_transform(X[categorical_features]).toarray()
scaler = StandardScaler()
X_numeric = scaler.fit_transform(X[numeric_features])

# Combine processed features
X_processed = np.hstack((X_categorical, X_numeric))

# Combine processed features with target
final_data = np.hstack((X_processed, y.reshape(-1, 1)))

# Get the feature names from one-hot encoding
encoded_feature_names = encoder.get_feature_names_out(categorical_features)
all_feature_names = np.hstack((encoded_feature_names, numeric_features, ['target']))

# Convert to DataFrame
df_final = pd.DataFrame(final_data, columns=all_feature_names)

# Save to CSV
df_final.to_csv('preprocessed_breast_cancer_data.csv', index=False)


# Split the dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42)

# Initialize hyperparameters
input_size = X_train.shape[1]
hidden_size = 10  # Hidden layer with 50 neurons
output_size = 1  # Binary classification
learning_rate = 0.001
epochs = 7000

# Initialize weights and biases
np.random.seed(42)  # For reproducibility
W1 = np.random.randn(input_size, hidden_size)
b1 = np.random.randn(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.random.randn(output_size)

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)
    return Z1, A1, Z2, A2

# Backward propagation
def backward_propagation(X, y, Z1, A1, Z2, A2, W1, W2, b1, b2, learning_rate):
    m = y.shape[0]

    # Calculate the error
    dZ2 = A2 - y.reshape(-1, 1)  # Output error
    dW2 = np.dot(A1.T, dZ2) / m  # Gradient for W2
    db2 = np.sum(dZ2, axis=0) / m  # Gradient for b2

    dA1 = np.dot(dZ2, W2.T)  # Error propagated back to hidden layer
    dZ1 = dA1 * sigmoid_derivative(Z1)
    dW1 = np.dot(X.T, dZ1) / m  # Gradient for W1
    db1 = np.sum(dZ1, axis=0) / m  # Gradient for b1

    # Update weights and biases
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    return W1, b1, W2, b2

# Error calculation (Mean Squared Error)
def calculate_error(y_true, y_pred):
    return np.mean(0.5 * (y_true - y_pred)**2)

# Training the neural network
errors = []
weights_over_time = []

for epoch in range(epochs):
    Z1, A1, Z2, A2 = forward_propagation(X_train, W1, b1, W2, b2)

    # Calculate error
    error = calculate_error(y_train, A2)
    errors.append(error)

    # Store weights for visualization
    weights_over_time.append(W1.copy())

    # Perform backpropagation and update weights and biases
    W1, b1, W2, b2 = backward_propagation(X_train, y_train, Z1, A1, Z2, A2, W1, W2, b1, b2, learning_rate)

    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Error: {error:.4f}')

# Plot error vs. epochs
plt.figure(figsize=(8, 6))
plt.plot(errors)
plt.title('Error vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error (Loss)')
plt.grid(True)
plt.show()

# Predictions on the test set
_, _, _, y_pred_test = forward_propagation(X_test, W1, b1, W2, b2)
y_pred_test = np.where(y_pred_test > 0.5, 1, 0)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred_test)
conf_matrix = confusion_matrix(y_test, y_pred_test)

# Confusion matrix plot
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Recurrence', 'Recurrence'], yticklabels=['No Recurrence', 'Recurrence'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix)

# Visualizing changes in weights
weights_over_time = np.array(weights_over_time)

# Visualizing how the weights of the first few neurons change over epochs
plt.figure(figsize=(10, 6))
for i in range(5):  # Visualize the first 5 neurons' weights over time
    plt.plot(weights_over_time[:, i, 0], label=f'Neuron {i+1}')
plt.title('Weight Changes Over Epochs for First 5 Neurons')
plt.xlabel('Epochs')
plt.ylabel('Weight Value')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score

# Fetch dataset
breast_cancer = fetch_ucirepo(id=14)

# Extract features and target
X = breast_cancer.data.features
y = breast_cancer.data.targets

# One-hot encode the target for binary classification
y = np.where(y == 'no-recurrence-events', 0, 1)

# Preprocessing (One-hot encode categorical features, standardize numeric)
categorical_features = ['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'breast', 'breast-quad', 'irradiat']
numeric_features = ['deg-malig']

encoder = OneHotEncoder(handle_unknown='ignore')
X_categorical = encoder.fit_transform(X[categorical_features]).toarray()
scaler = StandardScaler()
X_numeric = scaler.fit_transform(X[numeric_features])

# Combine processed features
X_processed = np.hstack((X_categorical, X_numeric))

# Initialize hyperparameters
input_size = X_processed.shape[1]
hidden_size = 10  # Hidden layer with 10 neurons
output_size = 1  # Binary classification
learning_rate = 0.001
epochs = 10000

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)
    return Z1, A1, Z2, A2

# Backward propagation
def backward_propagation(X, y, Z1, A1, Z2, A2, W1, W2, b1, b2, learning_rate):
    m = y.shape[0]

    # Calculate the error
    dZ2 = A2 - y.reshape(-1, 1)  # Output error
    dW2 = np.dot(A1.T, dZ2) / m  # Gradient for W2
    db2 = np.sum(dZ2, axis=0) / m  # Gradient for b2

    dA1 = np.dot(dZ2, W2.T)  # Error propagated back to hidden layer
    dZ1 = dA1 * sigmoid_derivative(Z1)
    dW1 = np.dot(X.T, dZ1) / m  # Gradient for W1
    db1 = np.sum(dZ1, axis=0) / m  # Gradient for b1

    # Update weights and biases
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    return W1, b1, W2, b2

# Error calculation (Mean Squared Error)
def calculate_error(y_true, y_pred):
    return np.mean(0.5 * (y_true - y_pred)**2)

# Cross-validation using StratifiedKFold
n_splits = 5
kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store fold results
accuracy_per_fold = []
errors_per_fold = []
conf_matrices_per_fold = []

for fold, (train_index, test_index) in enumerate(kf.split(X_processed, y)):
    print(f"Training on fold {fold+1}/{n_splits}...")

    # Split data
    X_train, X_test = X_processed[train_index], X_processed[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Initialize weights and biases (reset for each fold)
    np.random.seed(42)
    W1 = np.random.randn(input_size, hidden_size)
    b1 = np.random.randn(hidden_size)
    W2 = np.random.randn(hidden_size, output_size)
    b2 = np.random.randn(output_size)

    # Training the neural network
    errors = []
    for epoch in range(epochs):
        Z1, A1, Z2, A2 = forward_propagation(X_train, W1, b1, W2, b2)

        # Calculate error
        error = calculate_error(y_train, A2)
        errors.append(error)

        # Perform backpropagation and update weights and biases
        W1, b1, W2, b2 = backward_propagation(X_train, y_train, Z1, A1, Z2, A2, W1, W2, b1, b2, learning_rate)

    # Store errors for this fold
    errors_per_fold.append(errors)

    # Predictions on the test set
    _, _, _, y_pred_test = forward_propagation(X_test, W1, b1, W2, b2)
    y_pred_test = np.where(y_pred_test > 0.5, 1, 0)

    # Evaluate model performance on this fold
    accuracy = accuracy_score(y_test, y_pred_test)
    conf_matrix = confusion_matrix(y_test, y_pred_test)

    accuracy_per_fold.append(accuracy)
    conf_matrices_per_fold.append(conf_matrix)

    print(f"Fold {fold+1}, Accuracy: {accuracy * 100:.2f}%")
    print(f"Confusion Matrix for fold {fold+1}:")
    print(conf_matrix)

# Final Results after Cross-Validation
mean_accuracy = np.mean(accuracy_per_fold)
std_accuracy = np.std(accuracy_per_fold)

print(f"\nAverage Accuracy over {n_splits} folds: {mean_accuracy * 100:.2f}%")
print(f"Standard Deviation in Accuracy: {std_accuracy * 100:.2f}%")

# Plot error vs. epochs for each fold
plt.figure(figsize=(8, 6))
for i, error_list in enumerate(errors_per_fold):
    plt.plot(error_list, label=f'Fold {i+1}')
plt.title('Error vs. Epochs for Each Fold')
plt.xlabel('Epochs')
plt.ylabel('Error (Loss)')
plt.legend()
plt.grid(True)
plt.show()

# Average confusion matrix
avg_conf_matrix = np.mean(conf_matrices_per_fold, axis=0)

# Confusion matrix plot
plt.figure(figsize=(6, 5))
sns.heatmap(avg_conf_matrix, annot=True, fmt='.2f', cmap='Blues', cbar=False, xticklabels=['No Recurrence', 'Recurrence'], yticklabels=['No Recurrence', 'Recurrence'])
plt.title('Average Confusion Matrix Over Folds')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

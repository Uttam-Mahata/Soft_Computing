# Fuzzy Logic

### Let A(x) and B(x) be two fuzzy sets of speed limiton a highway, defined by the following membership functions

#### Fuzzy set A: “Low speed limit”

μ<sub>A(x)</sub> = {1, if x < 30 (low speed)
<br>

$\frac{x−30}{50−30}$, if 30 < x < 50 (moderately low speed)
<br>

$\frac{x−50}{70−50}$, if 50 < x < 70 (increasing speed)

<br>
0, if x > 70 (not low speed)}

#### Fuzzy set B: “High speed limit”

μ<sub>B(x)</sub> = {0, if x < 60 (Not high speed)
<br>

$\frac{x−60}{80−60}$, if 60 < x < 80 (moderately high speed)
<br>

$\frac{x−80}{100−80}$, if 80 < x < 100 (increasing speed)
<br>
1, if x > 100 (high speed)}

```python
#Import Libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python

def mu_A(x):
    if x < 30:
        return 1
    elif 30 < x < 50:
        return (x - 30) / (50 - 30)
    elif 50 < x < 70:
        return (x - 50) / (70 - 50)
    else:
        return 0
    



x = np.linspace(0, 100,100)
y = [mu_A(i) for i in x]
plt.plot(x, y)
plt.title('Fuzzy set A')
plt.show()


    
```

```python
def mu_B(x):
    if x < 60:
        return 0
    elif 60 < x < 80:
        return (x - 60) / (80 - 60)
    elif 80 < x < 100:
        return (x - 80) / (100 - 80)
    else:
        return 1
    


x = np.linspace(0, 100, 100)
y = [mu_B(i) for i in x]
plt.plot(x, y)
plt.title('Fuzzy set B')
plt.show()
```

```python


x = np.linspace(0, 100, 100)
y = [mu_A(i) for i in x]
plt.plot(x, y, label='A(x)')
y = [mu_B(i) for i in x]
plt.plot(x, y, label='B(x)')
plt.legend()
plt.show()
```

<p>Now, need to perform the following operations to obtain the resultant set and plot the same:<p>
i. Union - (A U B)x = max(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)
<br>

ii. Intersection - $(A \cap B)x$ = min(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)
<br>

iii. Complement of both sets - A'(x) =  (1 − μ<sub>A(x)</sub>), B'(x) =  (1 − μ<sub>B(x)</sub>)
<br>
iv. Difference of the two sets – (A − B)x = (μ<sub>A(x)</sub> − μ<sub>B(x)</sub>)

#### i. Union - (A U B)x = max(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)

```python
def union(x):
    return max(mu_A(x), mu_B(x))

#Plot the function


x = np.linspace(0, 100, 100)
y = [union(i) for i in x]
plt.plot(x, y)
plt.title('Union of A and B')
plt.show()
```

#### ii. Intersection - $(A \cap B)x$ = min(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)

```python
def intersection(x):
    return min(mu_A(x), mu_B(x))

#Plot the function


x = np.linspace(0, 100, 100)
y = [intersection(i) for i in x]
plt.plot(x, y)
plt.title('Intersection of A and B')
plt.show()
```

#### iii. Complement of both sets - A'(x) =  (1 − μ<sub>A(x)</sub>), B'(x) =  (1 − μ<sub>B(x)</sub>)

```python
#Complement of A
def complement_mu_A(x):
    return 1 - mu_A(x)

#Plot the function


x = np.linspace(0, 100, 100)
y = [complement_mu_A(i) for i in x]
plt.plot(x, y)
plt.title('Complement of A')
plt.show()
```

```python
#Complement of B
def complement_mu_B(x):
    return 1 - mu_B(x)

#Plot the function


x = np.linspace(0, 100, 100)
y = [complement_mu_B(i) for i in x]
plt.plot(x, y)
plt.title('Complement of B')
plt.show()
```

```python
#Difference of A and B
def difference_mu_A_B(x):
    return max(0, mu_A(x) - mu_B(x))

#Plot the function


x = np.linspace(0, 100, 100)
y = [difference_mu_A_B(i) for i in x]
plt.plot(x, y)
plt.title('Difference of A and B')
plt.show()
```

### Let C(x) and D(x) be two fuzzy sets of temperature (Celsius) in a city, defined by the following membership functions:

#### Fuzzy set C: “Cold”

μ<sub>C(x)</sub> = {0, if x <− 10 (extremely cold)
<br>

$\frac{x−(−10)}{0−(−10)}$, if − 10 < x < 0(very cold)
<br>
1, if0 < x < 5(cold)
<br>
0, if x > 5 (not cold)}


#### Fuzzy set D: “Warm”

μ<sub>D(x)</sub> = {0, if x < 25 (Not warm)
<br>

$\frac{x−25}{35−25}$, if 25 < x < 35(moderately warm)
<br>
1, if 35 < x < 40(warm)
0, if x > 40 (extremely warm)}


```python
def mu_C(x):
    if x < -10:
        return 0 # Extermely Cold
    elif -10 < x < 0:
        return (x + 10) / (0 + 10) # Very Cold
    elif 0 < x < 5:
        return 1 # Cold
    else:
        return 0 # Not Cold
    

def mu_D(x):
    if x < 25:
        return 0 # Not Warm
    elif 25 < x < 35:
        return (x - 25) / (35 - 25) # moderately warm
    elif 35 < x < 40:
        return 1 # Warm
    else:
        return 0 # Extremely Warm
    
    
    
x = np.linspace(-20, 20, 100)
y = [mu_C(i) for i in x]
plt.plot(x, y)
plt.title('Fuzzy set C')
plt.show()


x = np.linspace(0, 50, 100)
y = [mu_D(i) for i in x]
plt.plot(x, y)
plt.title('Fuzzy set D')
plt.show()

# Prepare a Dataframe for R : C x D
import pandas as pd
data = []
for i in x:
    data.append([mu_C(i), mu_D(i)])

df = pd.DataFrame(data, index=[''] columns=['C', 'D'])
df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def mu_C(x):
    if x <= -10:
        return 1  # Extremely Cold
    elif -10 < x <= 0:
        return (x + 10) / 10  # Very Cold
    elif 0 < x <= 5:
        return 1  # Cold
    else:
        return 0  # Not Cold
    

def mu_D(x):
    if x <= 25:
        return 0  # Not Warm
    elif 25 < x <= 35:
        return (x - 25) / 10  # Moderately Warm
    elif 35 < x <= 40:
        return 1  # Warm
    else:
        return 0  # Extremely Warm

# Generating the Cartesian Relation Matrix
x_values_C = np.linspace(-20, 10, 100)  # Values for C
x_values_D = np.linspace(20, 50, 100)   # Values for D

relation_matrix = np.zeros((4, 4))

for x in x_values_C:
    mu_c = mu_C(x)
    if mu_c == 1 and x <= -10:
        row = 0  # Extremely Cold
    elif 0 < mu_c < 1:
        row = 1  # Very Cold
    elif mu_c == 1 and 0 < x <= 5:
        row = 2  # Cold
    else:
        row = 3  # Not Cold

    for y in x_values_D:
        mu_d = mu_D(y)
        if mu_d == 0 and y <= 25:
            col = 0  # Not Warm
        elif 0 < mu_d < 1:
            col = 1  # Moderately Warm
        elif mu_d == 1 and 35 < y <= 40:
            col = 2  # Warm
        else:
            col = 3  # Extremely Warm

        relation_matrix[row, col] = max(relation_matrix[row, col], min(mu_c, mu_d))

# Creating DataFrame
df = pd.DataFrame(relation_matrix, 
                  index=['Extremely Cold', 'Very Cold', 'Cold', 'Not Cold'], 
                  columns=['Not Warm', 'Moderately Warm', 'Warm', 'Extremely Warm'])

# Display the DataFrame
df.head()
```

```python
import pandas as pd

x_values = np.arange(-10, 6, 3) 
z_values = np.arange(25, 41, 3)  


T = np.zeros((len(x_values), len(z_values)))




for i, x in enumerate(x_values):
    for j, z in enumerate(z_values):
        T[i, j] = max([min(mu_C(x), mu_D(z))])


table = pd.DataFrame(T, columns=[f'z={z}' for z in z_values], index=[f'x={x}' for x in x_values])

# Display the table
print("Max-Min Composition Table:")
print(table)


```

```python
#Max product composition
x_values = np.arange(-10, 5, 3)  
z_values = np.arange(25, 40, 3)  

T = np.zeros((len(x_values), len(z_values)))


for i, x in enumerate(x_values):
    for j, z in enumerate(z_values):
        T[i, j] = max([mu_C(x) * mu_D(z)])




headers_x = [f'x{i+1} = {x}' for x in x_values]
headers_z = [f'z{j+1} = {z}' for z in z_values]

table = pd.DataFrame(T, columns=headers_z, index=headers_x)


print("Max-Product Composition Table:")
print(table)
```

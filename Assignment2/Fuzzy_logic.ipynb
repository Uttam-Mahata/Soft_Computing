### Let A(x) and B(x) be two fuzzy sets of  limiton a highway, defined by the following membership functions

#### Fuzzy set A: “Low  limit”

μ<sub>A(x)</sub> = {1, if x < 30 (low )
<br>

$\frac{x−30}{50−30}$, if 30 < x < 50 (moderately low )
<br>

$\frac{x−50}{70−50}$, if 50 < x < 70 (increasing )

<br>
0, if x > 70 (not low )}

#### Fuzzy set B: “High  limit”

μ<sub>B(x)</sub> = {0, if x < 60 (Not high )
<br>

$\frac{x−60}{80−60}$, if 60 < x < 80 (moderately high )
<br>

$\frac{x−80}{100−80}$, if 80 < x < 100 (increasing )

i. Union - (A U B)x = max(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)
<br>

ii. Intersection - $(A \cap B)x$ = min(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)

<br>
iv. Difference of the two sets – (A − B)x = (μ<sub>A(x)</sub> − μ<sub>B(x)</sub>)

#### i. Union - (A U B)x = max(μ<sub>A(x)</sub>, μ<sub>B(x)</sub>)

### Let C(x) and D(x) be two fuzzy sets of temperature (Celsius) in a city, defined by the following membership functions:

#### Fuzzy set C: “Cold”

μ<sub>C(x)</sub> = {0, if x <− 10 (extremely cold)
<br>

$\frac{x−(−10)}{0−(−10)}$, if − 10 < x < 0(very cold)
<br>
1, if0 < x < 5(cold)
<br>
0, if x > 5 (not cold)}


#### Fuzzy set D: “Warm”

μ<sub>D(x)</sub> = {0, if x < 25 (Not warm)
<br>

$\frac{x−25}{35−25}$, if 25 < x < 35(moderately warm)
<br>
1, if 35 < x < 40(warm)
0, if x > 40 (extremely warm)}


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the membership function mu_C(x)
def mu_C(x):
    if x < -10:
        return 0  # Extremely Cold
    elif -10 <= x < 0:
        return (x + 10) / 10  # Very Cold
    elif 0 <= x < 5:
        return 1  # Cold
    else:
        return 0  # Not Cold

# Define the membership function mu_D(x)
def mu_D(x):
    if x < 25:
        return 0  # Not Warm
    elif 25 <= x < 35:
        return (x - 25) / 10  # Moderately Warm
    elif 35 <= x < 40:
        return 1  # Warm
    else:
        return 0  # Extremely Warm

# Generate x values for plotting fuzzy set C
x_values_C = np.linspace(-20, 20, 500)
mu_C_values = [mu_C(x) for x in x_values_C]

# Scatter plot with different colors for each region in fuzzy set C
plt.scatter(x_values_C, mu_C_values, c=['blue' if x < -10 
                                        else 'cyan' if -10 <= x < 0 
                                        else 'green' if 0 <= x < 5 
                                        else 'red' for x in x_values_C])

# Add labels, title, and grid for fuzzy set C
plt.title('Scatter Plot of Fuzzy Set C')
plt.xlabel('Temperature (x)')
plt.ylabel('Membership Value')
plt.grid(True)
plt.show()

# Generate x values for plotting fuzzy set D
x_values_D = np.linspace(0, 50, 500)
mu_D_values = [mu_D(x) for x in x_values_D]

# Scatter plot with different colors for each region in fuzzy set D
plt.scatter(x_values_D, mu_D_values, c=['gray' if x < 25 
                                        else 'orange' if 25 <= x < 35 
                                        else 'yellow' if 35 <= x < 40 
                                        else 'red' for x in x_values_D])

# Add labels, title, and grid for fuzzy set D
plt.title('Scatter Plot of Fuzzy Set D')
plt.xlabel('Temperature (x)')
plt.ylabel('Membership Value')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define membership functions
def mu_C(x):
    if x < -10:
        return 0  # Extremely Cold
    elif -10 <= x <= 0:
        return (x + 10) / 10  # Very Cold
    elif 0 < x <= 5:
        return 1  # Cold
    else:
        return 0  # Not Cold
    

def mu_D(x):
    if x < 25:
        return 0  # Not Warm
    elif 25 <= x <= 35:
        return (x - 25) / 10  # Moderately Warm
    elif 35 < x <= 40:
        return 1  # Warm
    else:
        return 0  # Extremely Warm

# Step 2: Compute Fuzzy Relation R(C × D)
x_values_C = np.linspace(-20, 20, 100)
x_values_D = np.linspace(0, 50, 100)

R = np.zeros((len(x_values_C), len(x_values_D)))

for i, x_C in enumerate(x_values_C):
    for j, x_D in enumerate(x_values_D):
        R[i, j] = min(mu_C(x_C), mu_D(x_D))

# Step 3: Perform Fuzzy Set Composition (Max-Min and Max-Product)
C_comp = np.zeros(len(x_values_D))
D_comp = np.zeros(len(x_values_C))

# Max-Min Composition
for j in range(len(x_values_D)):
    C_comp[j] = max([min(R[i, j], mu_C(x_values_C[i])) for i in range(len(x_values_C))])

for i in range(len(x_values_C)):
    D_comp[i] = max([min(R[i, j], mu_D(x_values_D[j])) for j in range(len(x_values_D))])


# Plotting the results of the compositions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_values_D, C_comp, label='C o (C × D)')
plt.title('Max-Min Composition C o (C × D)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_values_C, D_comp, label='D o (C × D)')
plt.title('Max-Min Composition D o (C × D)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

plt.show()
```


### Max-Min Composition
The Max-Min composition of fuzzy sets is defined as:

$
C \circ (C \times D) = \max_{x_C} \left[ \min(\mu_C(x_C), R(x_C, x_D)) \right]
$

This operation is interpreted as the degree to which each element in the fuzzy set `D` (or `C`) is related to the fuzzy set `C` (or `D`), considering the fuzzy relation `R`.

### Max-Product Composition
The Max-Product composition is similar but uses multiplication instead of the minimum:

$
C \circ (C \times D) = \max_{x_C} \left[ \mu_C(x_C) \times R(x_C, x_D) \right]
$

This operation represents a more weighted approach, where the degree of membership in `D` (or `C`) is scaled by the strength of the fuzzy relation.

### Interpretation of the Compositions

1. **Max-Min Composition Results**:
   - **For `C o (C × D)`**: This composition shows how "cold" the temperatures are across the range of `D`. The result gives the maximum membership value of `C` across all the temperatures that are related to each temperature in `D` (Warm). 
     - For example, if the result shows a high value for a temperature in `D`, it means that this temperature has a strong relationship with being "cold."
   - **For `D o (C × D)`**: This shows how "warm" the temperatures are across the range of `C`. The result gives the maximum membership value of `D` across all the temperatures that are related to each temperature in `C` (Cold).
     - If a temperature in `C` has a high value in the result, it suggests that it also has a significant degree of "warmth."

2. **Max-Product Composition Results**:
   - **For `C o (C × D)`**: The Max-Product composition gives a more nuanced interpretation. Instead of just taking the minimum, it multiplies the degrees of membership, which can show more subtle relationships between the temperatures. A high result here indicates that a temperature in `D` is not just related to being "cold," but is strongly weighted towards that relationship.
   - **For `D o (C × D)`**: Similarly, this composition for `D` indicates how much a temperature in `C` is not only related to being "warm" but also how strongly this relationship holds.

### Trapezoidal Membership Function

The trapezoidal membership function is commonly used in fuzzy logic to represent a fuzzy set where the membership values increase, reach a plateau, and then decrease. It is particularly useful when we want to define a fuzzy number that has a range of values with full membership and gradual transitions to partial membership on either side.

#### Mathematical Expression

A trapezoidal membership function is typically defined by four parameters: \(a\), \(b\), \(c\), and \(d\). These parameters determine the shape of the trapezoid. The membership function $\mu(x) $is defined as:

$
\mu(x) =
\begin{cases}
0 & \text{if } x \leq a \\
\frac{x-a}{b-a} & \text{if } a < x \leq b \\
1 & \text{if } b < x \leq c \\
\frac{d-x}{d-c} & \text{if } c < x \leq d \\
0 & \text{if } x > d
\end{cases}
$

Where:
- \(a\) and \(d\) define the points where the membership degree starts rising from 0 and falls back to 0, respectively.
- \(b\) and \(c\) define the points where the membership degree is 1, meaning the value belongs fully to the fuzzy set.
- The intervals \([a, b]\) and \([c, d]\) represent the slopes of the trapezoid where the membership degree transitions between 0 and 1.

#### Interpretation

- **\(a\)**: The point at which the membership function starts to rise from 0.
- **\(b\)**: The point at which the membership function reaches 1.
- **\(c\)**: The point at which the membership function starts to fall from 1.
- **\(d\)**: The point at which the membership function returns to 0.

### Triangular Membership Function

The triangular membership function is one of the simplest and most commonly used types of membership functions in fuzzy logic. It is used to define fuzzy sets that are characterized by a linear increase to a peak and then a linear decrease, forming a triangle-like shape.

#### Mathematical Expression

A triangular membership function is defined by three parameters: \(a\), \(b\), and \(c\). The membership function $ \mu(x) $ is defined as:

$
\mu(x) =
\begin{cases}
0 & \text{if } x \leq a \\
\frac{x-a}{b-a} & \text{if } a < x \leq b \\
\frac{c-x}{c-b} & \text{if } b < x \leq c \\
0 & \text{if } x > c
\end{cases}
$

Where:
- \(a\) is the left endpoint where the membership degree starts to rise from 0.
- \(b\) is the peak of the triangle where the membership degree reaches 1, meaning the value belongs fully to the fuzzy set.
- \(c\) is the right endpoint where the membership degree decreases back to 0.

#### Interpretation

- **\(a\)**: The point at which the membership function starts to rise from 0.
- **\(b\)**: The point at which the membership function reaches its maximum value of 1.
- **\(c\)**: The point at which the membership function decreases back to 0.



### Gaussian Membership Function

The Gaussian membership function is a key concept in fuzzy logic, particularly valued for its smooth, bell-shaped curve. It is widely used to represent fuzzy sets where the membership of elements gradually decreases as they move away from a central value. This makes it especially useful for modeling concepts with a central tendency and no abrupt changes.

#### Mathematical Expression

The Gaussian membership function is mathematically defined by two parameters: \(c\) (the mean or center) and  $ \sigma $ (the standard deviation). The function $ \mu(x) $ is given by:

$
\mu(x) = \exp\left(-\frac{(x-c)^2}{2\sigma^2}\right)
$

Where:
- \(x\) is the input value.
- \(c\) is the center of the Gaussian curve, where the membership degree reaches its maximum (1).
- $ \sigma $ is the standard deviation, which controls the width of the curve.

#### Interpretation

- **\(c\)**: The center of the Gaussian function, representing the value with the highest membership degree (1), indicating full membership in the fuzzy set.
- $ \sigma $: The standard deviation, determining the spread of the curve. A smaller $ \sigma $ results in a narrower, steeper curve, while a larger $ \sigma $ produces a wider, more gradual curve.

#### Characteristics

- **Smoothness**: The Gaussian function is continuous and infinitely differentiable, which makes it ideal for situations where smooth transitions between membership grades are desired.
- **Symmetry**: The function is symmetric around the center \(c\), meaning the membership decreases equally on both sides as the distance from \(c\) increases.
- **No Sharp Boundaries**: Unlike other membership functions like triangular or trapezoidal, the Gaussian function has no sharp edges, making it better suited for modeling gradual changes in membership.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 8, 100)


def trapezoidal(x, a, b, c, d):
    return np.maximum(0, np.minimum((x-a)/(b-a), np.minimum(1, (d-x)/(d-c))))


def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x-a)/(b-a), (c-x)/(c-b)))


def gaussian(x, c, sigma):
    return np.exp(-0.5 * ((x - c) / sigma) ** 2)


a_trap, b_trap, c_trap, d_trap = 2, 3, 5, 6
a_tri, b_tri, c_tri = 2, 4, 6
c_gauss, sigma_gauss = 4, 1


trap_values = trapezoidal(x, a_trap, b_trap, c_trap, d_trap)
tri_values = triangular(x, a_tri, b_tri, c_tri)
gauss_values = gaussian(x, c_gauss, sigma_gauss)

# Plotting the membership functions
plt.figure(figsize=(15, 5))

# Plot Trapezoidal Membership Function
plt.subplot(1, 3, 1)
plt.plot(x, trap_values, label='Trapezoidal', color='blue')
plt.title('Trapezoidal Membership Function')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.ylim(0, 1.1)

# Plot Triangular Membership Function
plt.subplot(1, 3, 2)
plt.plot(x, tri_values, label='Triangular', color='green')
plt.title('Triangular Membership Function')
plt.xlabel('x')
plt.grid(True)
plt.ylim(0, 1.1)

# Plot Gaussian Membership Function
plt.subplot(1, 3, 3)
plt.plot(x, gauss_values, label='Gaussian', color='red')
plt.title('Gaussian Membership Function')
plt.xlabel('x')
plt.grid(True)
plt.ylim(0, 1.1)

plt.tight_layout()
plt.show()
```

#### Design a Mamdani Fuzzy Inference System to control the FAN-SPEED of a furnace by inputting TEMPERATURE of a thermostat of a household.

#### Frame the If-Then rules using Linguistic variables input TEMPERATURE and output FAN-SPEED.
##### Sample rule:
- “IF TEMPERATURE is 85, THEN increase furnace FAN-SPEED to 300 RPM ”
- (i) Write fuzzy rules (as many as possible) using the linguistic variable TEMPERATURE and FAN-SPEED. Consider Fuzzy variables for TEMPERATURE like “Risky”, “Average”, “Excellent” and FAN-SPEED like “Slow”, “High”, “Moderate” etc.
- (ii) Fuzzify the input and output fuzzy variables with appropriate membership functions.
- (iii) Aggregate the rules using Mamdani model
- (iv) Apply Centroid defuzzification method to calculate the crisp value of FAN-SPEED.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the universe of discourse for temperature and fan speed
x_temp = np.arange(200, 1000, 1)  
x_fan = np.arange(500, 3000, 1)

# Define fuzzy membership functions for Temperature
def trimf(x, a, b, c):
    return np.maximum(0, np.minimum((x-a)/(b-a), (c-x)/(c-b)))

# Define fuzzy membership functions for Fan Speed
def trapmf(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x-a)/(b-a), 1), (d-x)/(d-c)))

# Temperature membership functions
temp_risky = trimf(x_temp, 60, 70, 80)
temp_average = trimf(x_temp, 75, 85, 95)
temp_excellent = trimf(x_temp, 90, 100, 110)

# Fan Speed membership functions
fan_slow = trapmf(x_fan, 100, 150, 200, 250)
fan_moderate = trimf(x_fan, 200, 300, 400)
fan_high = trapmf(x_fan, 350, 400, 450, 500)

# Visualize these membership functions
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(8, 10))

ax1.plot(x_temp, temp_risky, 'r', label='Risky')
ax1.plot(x_temp, temp_average, 'g', label='Average')
ax1.plot(x_temp, temp_excellent, 'b', label='Excellent')
ax1.set_title('Temperature Membership Functions')
ax1.legend()

ax2.plot(x_fan, fan_slow, 'r', linestyle='--', label='Slow')
ax2.plot(x_fan, fan_moderate, 'g', linestyle='--', label='Moderate')
ax2.plot(x_fan, fan_high, 'b', linestyle='--', label='High')
ax2.set_title('Fan Speed Membership Functions')
ax2.legend()

plt.tight_layout()
plt.show()

# Example input temperature
temp_input = 80

# Fuzzify the input temperature
def fuzzify(x, mf):
    return np.interp(x, x_temp, mf)

temp_risky_level = fuzzify(temp_input, temp_risky)
temp_average_level = fuzzify(temp_input, temp_average)
temp_excellent_level = fuzzify(temp_input, temp_excellent)

# Apply fuzzy rules using Mamdani method
rule1 = np.fmin(temp_risky_level, fan_high)
rule2 = np.fmin(temp_average_level, fan_moderate)
rule3 = np.fmin(temp_excellent_level, fan_slow)

# Combine the rules (Aggregation)
aggregated = np.fmax(rule1, np.fmax(rule2, rule3))

# Defuzzify the result using the Centroid method
def defuzz(x, mf):
    return np.sum(x * mf) / np.sum(mf)

fan_speed = defuzz(x_fan, aggregated)
fan_speed_activation = np.interp(fan_speed, x_fan, aggregated)

# Visualize the aggregated output and defuzzified result
fig, ax0 = plt.subplots(figsize=(8, 5))

ax0.plot(x_fan, fan_slow, 'r', linestyle='--', label='Slow')
ax0.plot(x_fan, fan_moderate, 'g', linestyle='--', label='Moderate')
ax0.plot(x_fan, fan_high, 'b', linestyle='--', label='High')
ax0.fill_between(x_fan, 0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([fan_speed, fan_speed], [0, fan_speed_activation], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Aggregated Membership and Result (Fan Speed)')
ax0.legend()

plt.tight_layout()
plt.show()

# Print the fuzzy input levels and final crisp fan speed
print(f"Input Temperature: {temp_input}")
print(f"Risky level: {temp_risky_level}")
print(f"Average level: {temp_average_level}")
print(f"Excellent level: {temp_excellent_level}")
print(f"Defuzzified Fan Speed: {fan_speed:.2f} RPM")
```

### Fuzzy Control System for Fan Speed Based on Temperature

In this exercise, need to  create a fuzzy control system where the input is the **TEMPERATURE** and the output is the **FAN-SPEED**. We’ll define linguistic variables for both, write fuzzy rules, fuzzify the inputs, aggregate the rules using the Mamdani model, and finally apply the centroid defuzzification method to calculate a crisp value for the fan speed.

---

#### (i) Fuzzy Rules

We'll define fuzzy rules using linguistic variables for **TEMPERATURE** and **FAN-SPEED**. The possible fuzzy sets (or labels) for each variable are as follows:

- **TEMPERATURE**: "Risky", "Average", "Excellent"
- **FAN-SPEED**: "Slow", "Moderate", "High"

**Sample Fuzzy Rules:**

1. **Rule 1**: IF TEMPERATURE is "Risky", THEN FAN-SPEED is "High".
2. **Rule 2**: IF TEMPERATURE is "Average", THEN FAN-SPEED is "Moderate".
3. **Rule 3**: IF TEMPERATURE is "Excellent", THEN FAN-SPEED is "Slow".
4. **Rule 4**: IF TEMPERATURE is "Risky" AND "Average", THEN FAN-SPEED is "Moderate".
5. **Rule 5**: IF TEMPERATURE is "Risky" AND "Excellent", THEN FAN-SPEED is "Moderate".
6. **Rule 6**: IF TEMPERATURE is "Average" AND "Excellent", THEN FAN-SPEED is "Moderate".
7. **Rule 7**: IF TEMPERATURE is "Average" AND NOT "Excellent", THEN FAN-SPEED is "High".
8. **Rule 8**: IF TEMPERATURE is "Risky" OR "Average", THEN FAN-SPEED is "High".

---

#### (ii) Fuzzification

Fuzzification is the process of mapping crisp input values to fuzzy sets with corresponding membership values. Here, we'll define the membership functions for **TEMPERATURE** and **FAN-SPEED**.

**Membership Functions:**

- **TEMPERATURE**
  - **"Risky"**: Triangular with \(a=60\), \(b=70\), \(c=80\)
  - **"Average"**: Triangular with \(a=75\), \(b=85\), \(c=95\)
  - **"Excellent"**: Triangular with \(a=90\), \(b=100\), \(c=110\)

- **FAN-SPEED**
  - **"Slow"**: Trapezoidal with \(a=100\), \(b=150\), \(c=200\), \(d=250\)
  - **"Moderate"**: Triangular with \(a=200\), \(b=300\), \(c=400\)
  - **"High"**: Trapezoidal with \(a=350\), \(b=400\), \(c=450\), \(d=500\)



#### (iii) Rule Aggregation Using Mamdani Model

Using the Mamdani model, we aggregate the output from all rules by taking the minimum of the membership values for each rule's antecedent (the "IF" part) and using this to clip the corresponding consequent (the "THEN" part).

**Example Aggregation:**

- **Rule 1**: IF TEMPERATURE is "Risky" (0.4), THEN FAN-SPEED is "High" → FAN-SPEED = "High" clipped at 0.4.
- **Rule 2**: IF TEMPERATURE is "Average" (0.7), THEN FAN-SPEED is "Moderate" → FAN-SPEED = "Moderate" clipped at 0.7.
- **Rule 3**: IF TEMPERATURE is "Excellent" (0.2), THEN FAN-SPEED is "Slow" → FAN-SPEED = "Slow" clipped at 0.2.

The overall fuzzy output will be the union of the clipped membership functions for "High", "Moderate", and "Slow" FAN-SPEEDs.

---

#### (iv) Defuzzification Using Centroid Method

Defuzzification is the process of converting the aggregated fuzzy output back into a crisp value. The **centroid method** (or center of gravity) is commonly used and calculates the crisp output by finding the center of mass of the aggregated fuzzy set.

The centroid $ z^* $ is calculated as:

$
z^* = \frac{\int z \cdot \mu(z) \, dz}{\int \mu(z) \, dz}
$

Where:
- $ \mu(z) $ is the aggregated membership function.
- The numerator represents the moment about the origin, and the denominator represents the total area under the curve.

**Final Step:**

Applied the centroid formula to the aggregated membership function to obtain the crisp FAN-SPEED.

```python
import numpy as np
import matplotlib.pyplot as plt

x_temp = np.arange(200, 1000, 1)  
x_fan = np.arange(500, 3000, 1)   


def trimf(x, a, b, c):
    return np.maximum(0, np.minimum((x-a)/(b-a), (c-x)/(c-b)))

# Define fuzzy membership functions for Fan Speed
def trapmf(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x-a)/(b-a), 1), (d-x)/(d-c)))

# Temperature membership functions
temp_risky = trimf(x_temp, 200, 300, 400)  # Modified parameters
temp_average = trimf(x_temp, 350, 500, 650)  # Modified parameters
temp_excellent = trimf(x_temp, 600, 800, 1000)  # Modified parameters

# Fan Speed membership functions
fan_slow = trapmf(x_fan, 500, 800, 1300, 1600)  # Modified parameters
fan_moderate = trimf(x_fan, 1200, 1700, 2200)  # Modified parameters
fan_high = trapmf(x_fan, 2000, 2400, 2700, 3000)  # Modified parameters

# Visualize these membership functions
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(8, 10))

ax1.plot(x_temp, temp_risky, 'r', label='Risky')
ax1.plot(x_temp, temp_average, 'g', label='Average')
ax1.plot(x_temp, temp_excellent, 'b', label='Excellent')
ax1.set_title('Temperature Membership Functions')
ax1.legend()

ax2.plot(x_fan, fan_slow, 'r', linestyle='--', label='Slow')
ax2.plot(x_fan, fan_moderate, 'g', linestyle='--', label='Moderate')
ax2.plot(x_fan, fan_high, 'b', linestyle='--', label='High')
ax2.set_title('Fan Speed Membership Functions')
ax2.legend()

plt.tight_layout()
plt.show()

# Example input temperature
temp_input = 250

# Fuzzify the input temperature
def fuzzify(x, mf):
    return np.interp(x, x_temp, mf)

temp_risky_level = fuzzify(temp_input, temp_risky)
temp_average_level = fuzzify(temp_input, temp_average)
temp_excellent_level = fuzzify(temp_input, temp_excellent)

# Apply fuzzy rules using Mamdani method
rule1 = np.fmin(temp_risky_level, fan_high)
rule2 = np.fmin(temp_average_level, fan_moderate)
rule3 = np.fmin(temp_excellent_level, fan_slow)

# Combine the rules (Aggregation)
aggregated = np.fmax(rule1, np.fmax(rule2, rule3))

# Defuzzify the result using the Centroid method
def defuzz(x, mf):
    return np.sum(x * mf) / np.sum(mf)

fan_speed = defuzz(x_fan, aggregated)
fan_speed_activation = np.interp(fan_speed, x_fan, aggregated)

# Visualize the aggregated output and defuzzified result
fig, ax0 = plt.subplots(figsize=(8, 5))

ax0.plot(x_fan, fan_slow, 'r', linestyle='--', label='Slow')
ax0.plot(x_fan, fan_moderate, 'g', linestyle='--', label='Moderate')
ax0.plot(x_fan, fan_high, 'b', linestyle='--', label='High')
ax0.fill_between(x_fan, 0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([fan_speed, fan_speed], [0, fan_speed_activation], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Aggregated Membership and Result (Fan Speed)')
ax0.legend()

plt.tight_layout()
plt.show()

# Print the fuzzy input levels and final crisp fan speed
print(f"Input Temperature: {temp_input}")
print(f"Risky level: {temp_risky_level}")
print(f"Average level: {temp_average_level}")
print(f"Excellent level: {temp_excellent_level}")
print(f"Defuzzified Fan Speed: {fan_speed:.2f} RPM")
```
